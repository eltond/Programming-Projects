###2.1 Begin by writing a function to compute the Root Mean Squared Error for a list of numbers

You can find the sqrt function in the Numpy package. Furthermore the details of RMSE can be found on [Wikipedia](http://en.wikipedia.org/wiki/Root-mean-square_deviation)



````python
"""
Function
--------
compute_rmse

Given two arrays, one of actual values and one of predicted values,
compute the Roote Mean Squared Error

Parameters
----------
predictions : array
    Array of numerical values corresponding to predictions for each of the N observations

yvalues : array
    Array of numerical values corresponding to the actual values for each of the N observations

Returns
-------
rmse : int
    Root Mean Squared Error of the prediction

Example
-------
>>> print compute_rmse((2,2,3),(0,2,6)
2.16
"""
import math;
def compute_rmse(predictions, yvalues):
    #your code here  
    sqsum=0.0
    rmse =0.0
  
    for i in range(0, len(yvalues)):
       sqsum = sqsum + np.power(predictions[i]-yvalues[i],2)

    rmse = np.sqrt(sqsum/len(predictions))
    return rmse

````

###2.4 Nearest Neighbors: Distance function
Let's try and build a machine learning algorithm to beat the "Average Value" baseline that you computed above.  Soon you will implement the Nearest Neighbor algorithm, but first you need to create a distance metric to measure the distance (and similarity) between two instances.  Write a generic function to compute the L-Norm distance (called the [*p*-norm](http://en.wikipedia.org/wiki/Norm_(mathematics) distance on Wikipedia)

````python
"""
Function
--------
distance

Given two instances and a value for L, return the L-Norm distance between them

Parameters
----------
x1, x2 : array
    Array of numerical values corresponding to predictions for each of the N observations

L: int
    Value of L to use in computing distances

Returns
-------
dist : int
    The L-norm distance between instances

Example
-------
>>> print distance((3,4),(5,9),1)
7

"""
def distance(x1, x2, L):
    #your code here
    sum=0;
    for i in range(0,len(x2)):
        sum=sum + np.power (x2[i] - x1[i],L)
    
    dist = np.power (sum, 1.0/float(L))
    
    return dist
````

###2.8 K-Nearest Neighbors Algorithm

Implement the K-Nearest Neighbors algorithm.  Using 10-fold cross validation, report the RMSE for K=3 and the running time of the algorithm.

````python
"""
# train_data: dataset for training the model
# medians_train: MEDV of training data
# test_data: dataset for testing data
# medians_obs_test: observed MEDV values of trained data
# feat_vect: vector of desired features
# L: value of L to be passed to distance fucntion
# K: value of K in K-NN
"""
def knn(train_data,medians_train,test_data,medians_obs_test,feat_vector, L, K):
    start_time = time.time()
    minDistanceVector=np.zeros(len(train_data))
    minDistMedv=0;
    
    predicted_medians=np.zeros(len(medians_obs_test))
    
    train_data_vect=np.zeros(len(feat_vector))
    test_data_vect=np.zeros(len(feat_vector))
    index_vect=0
      
    for i in range (0, len(test_data)):
        minDist = 999999
        #build up the feature vector of test data
        index_vect=0
        for k in feat_vector:
                    test_data_vect[index_vect]= test_data [i,k]
                    index_vect=index_vect + 1
        for j in range (0 , len(train_data)):
                # build up the feature vector of training data
                index_vect = 0
                for k in feat_vector:
                    train_data_vect[index_vect]=train_data [j,k]
                    index_vect=index_vect + 1
                
                # pass the test vector and train vector to the function for calculating distance
                dist = distance(test_data_vect,train_data_vect,L)
                minDistanceVector[j]=dist
        predicted_medians[i] = getMeanMedian(minDistanceVector,medians_train,K)
            
    rmse = compute_rmse (predicted_medians,medians_obs_test)
            

    #print "Time taken: " + str(round(time.time() - start_time,2)) + " seconds"
    return rmse
"""
# minDistanceVector: vector of all the distance values
# train_target: vector of MEDV values of training data
# K: number of min values to be selected
# returns the mean of MEDV for values for K nearest neighbors
"""
def getMeanMedian(minDistanceVector,train_target,K):
    
    sum = 0
    min=999999
    min_index=0
    
    for i in range (0,K):
        min = 999999
        for j in range (0,len(minDistanceVector)):
            if ( minDistanceVector[j] < min):
                min = minDistanceVector[j] 
                min_index = j
        sum = sum + train_target[min_index]
        minDistanceVector[min_index] = 9999999
    
    
    return sum/float(K)

"""
This function performs cross validation with knn function with a specified value of K
"""
def tenFoldValidationK (feat_vect,nfolds,K):

    rmse_vals = np.zeros(10)
    i=0
    kf = KFold (len(target_train), n_folds = nfolds, indices = False)
    for train, test in kf:
        
        rmse = knn (bdata_train_norm[train], target_train[train],bdata_train_norm[test],target_train[test],feat_vect,2,K)
        rmse_vals[i] = rmse
        i=i+1
        
    return np.mean(rmse_vals)

start_time = time.time()
print tenFoldValidationK((0,6,7,9), 2,3)
print "Time taken: " + str(round(time.time() - start_time,2)) + " seconds"
````
